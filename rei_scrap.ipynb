```python
import httpx
from selectolax.parser import HTMLParser
```

```python
url="https://www.decathlon.in/camping/tents-and-shelters-22753?id=22753&type=c"
headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36"}
timeout = httpx.Timeout(60.0)

try:
    response = httpx.get(url=url,headers=headers, timeout=timeout)
    response.raise_for_status()  # Check for HTTP errors
    # Process the response
except httpx.RequestError as exc:
    print(f"An error occurred: {exc}")
except httpx.HTTPStatusError as exc:
    print(f"HTTP error occurred: {exc}")
```

```python
html=HTMLParser(response.text)
```

```python
print(html.css_first("title").text())
```

## 

```python
# Correct selector for multiple classes
products = html.css('div.ais-InfiniteHits ul li')
```

```python
# Loop through and print the product names
# Loop through and print the product names
products
```

```python
# Loop through each product and extract the name and price
for product in products:
    name = product.css_first('p.capitalize.text-14').text(strip=True)
    price = product.css_first('span.text-14.font-semibold').text(strip=True)
    print(f"Product Name: {name}, Price: {price}")
```
