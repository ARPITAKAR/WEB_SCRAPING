```python
import httpx
from selectolax.parser import HTMLParser
import pandas as pd

def get_html():
    # URL of the page you want to scrape
    url = 'https://firstlens.in/collections/fashionable-sunglasses'  # Replace with the actual URL

    # Fetch the HTML content
    response = httpx.get(url)
    response.raise_for_status()  # Raise an exception for HTTP errors
    html = HTMLParser(response.text)
    return html

def parse_page(html):
    # Select all product-card elements
    product_cards = html.css('product-card')
    Sale_price = []
    Regular_price = []

    # Extract and print sale price and regular price
    for card in product_cards:
        # Find the price-list within the product-card
        price_list = card.css_first('price-list')
        
        if price_list:
            # Extract sale price
            sale_price_element = price_list.css_first('sale-price')
            sale_price = sale_price_element.text(strip=True) if sale_price_element else 'Not available'
            Sale_price.append(sale_price)
            
            # Extract regular price
            compare_at_price_element = price_list.css_first('compare-at-price')
            compare_at_price = compare_at_price_element.text(strip=True) if compare_at_price_element else 'Not available'
            Regular_price.append(compare_at_price)
            
    # Create a DataFrame from the lists
    df = pd.DataFrame({
        'Sale Price': Sale_price,
        'Regular Price': Regular_price
    })
    
    # Print or return the DataFrame
    print(df)

def main():
    html = get_html()
    parse_page(html)

if __name__ == "__main__":
    main()
```
