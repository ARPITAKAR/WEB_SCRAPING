```python
import httpx
from selectolax.parser import HTMLParser
```

```python
url="https://sassafras.in/collections/mens-t-shirt"
headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36"}
timeout = httpx.Timeout(60.0)

try:
    response = httpx.get(url=url,headers=headers, timeout=timeout)
    response.raise_for_status()  # Check for HTTP errors
    # Process the response
except httpx.RequestError as exc:
    print(f"An error occurred: {exc}")
except httpx.HTTPStatusError as exc:
    print(f"HTTP error occurred: {exc}")
html=HTMLParser(response.text)
```

```python
print(html.css_first("title").text())
```

```python
# Select all product items
product_items = html.css('product-item')
```

```python
product_items
```

```python
# Select all product items
product_items = html.css('product-item')
for item in product_items:
    # Extract product name
    name = item.css_first('a.product-item-meta__title').text(strip=True)
    
    # Extract sale price and regular price
    price_list = item.css_first('div.price-list')
    sale_price = price_list.css_first('span.price--highlight').text(strip=True).replace('Rs.', '').strip()
    regular_price = price_list.css_first('span.price--compare').text(strip=True).replace('Rs.', '').strip()
    
    print(f"Product Name: {name}")
    print(f"Sale Price: ₹ {sale_price}")
    print(f"Regular Price: ₹ {regular_price}")
```
